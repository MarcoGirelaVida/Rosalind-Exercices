```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parámetros de la distribución
media = 0
desviacion_estandar = 1
num_muestras = 1000

# Genera las muestras
muestras = np.random.normal(media, desviacion_estandar, num_muestras)

# Crea un DataFrame de pandas con las muestras
df = pd.DataFrame(muestras, columns=['Muestras'])
```

```python
# Crea un histograma de las muestras
df['Muestras'].plot.hist(bins=30, density=True)  # Utilizamos el argumento density=True para normalizar los datos

# Para una mejor visualización, podemos superponer la función de densidad de probabilidad teórica
from scipy.stats import norm

# Genera valores para el eje x
valores_x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 100)

# Calcula los valores y de la función de densidad de probabilidad normal
valores_y = norm.pdf(valores_x, media, desviacion_estandar)

# Dibuja la función de densidad de probabilidad
plt.plot(valores_x, valores_y)
plt.show()
```

```python
# Parámetros de la distribución
lam = 5
num_muestras = 1000

# Genera las muestras
muestras = np.random.poisson(lam, num_muestras)

# Crea un DataFrame de pandas con las muestras
poisson = pd.DataFrame(muestras, columns=['Muestras'])

# Crea un histograma de las muestras
poisson['Muestras'].plot.hist(bins=30, density=True)  # Utilizamos el argumento density=True para normalizar los datos

plt.show()
```

```python
# Parámetros de la distribución
n = 1000000000
p = 0.5
num_muestras = 10000

# Genera las muestras
muestras = np.random.binomial(n, p, num_muestras)

# Crea un DataFrame de pandas con las muestras
binomial = pd.DataFrame(muestras, columns=['Muestras'])

# Crea un histograma de las muestras
binomial['Muestras'].plot.hist()  # Utilizamos el argumento density=True para normalizar los datos

print("Media esperada: ", n*p) 
print("Desviacion tipica esperada: ", np.sqrt(n*p*(1-p)))
plt.show()
```
